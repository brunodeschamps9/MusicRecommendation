# **Recomendador de Músicas pelo Spotify**

Usando a base de dados pública do Spotify, esse projeto é feito com a finalidade criar um algoritmo capaz de recomendar músicas com base em todas as suas variáveis, como se encontra abaixo:


*   *Acousticness/Acústica*
*   *Danceability/Dançabilidade*
*   *Duration_ms/Duração*
*   *Energy/Energia*
*   *Instrumentalness/Instrumentalidade*
*   *Liveness/ Ao vivo*
*   *Popularity/Popularidade*
*   *Valence/Valência*




[Spotfy API](https://developer.spotify.com/documentation/web-api/reference/get-several-audio-features)

# **Análise dos Dados:**

```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados_totais.csv')
dados.head()
```

```python
dados = dados.drop(["explicit", "key", "mode"], axis=1)
```

```python
dados.shape
```

```python
dados.isna().sum()
```

# **Padronizando/Escalando os Dados:**

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(dtype=int)
coluna_ohe = ohe.fit_transform(dados[['artists']]).toarray()
dados2 = dados.drop('artists', axis=1)

dados_musicas_dummies = pd.concat([dados2, pd.DataFrame(coluna_ohe, columns=ohe.get_feature_names_out(['artists']))], axis=1)
dados_musicas_dummies
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

pca_pipeline = Pipeline([('scaler',StandardScaler()), ('PCA',PCA(n_components=0.7, random_state=999))])

music_embending_pca = pca_pipeline.fit_transform(dados_musicas_dummies.drop(['id','name','artists_song'], axis=1))
projection_m = pd.DataFrame(data=music_embending_pca)
projection_m
```

# **Criando o modelo de Clustering utilizando o KMeans:**

```python
from sklearn.cluster import KMeans

kmeans_pca_pipeline = KMeans(n_clusters=50, verbose=False, random_state=999)

kmeans_pca_pipeline.fit(projection_m)

dados['cluster_pca'] = kmeans_pca_pipeline.predict(projection_m)
projection_m['cluster_pca'] = kmeans_pca_pipeline.predict(projection_m)
projection_m['artist'] = dados['artists']
projection_m['song'] = dados['artists_song']
```

```python
projection_m.head()
```

```python
nome_musica = 'Juice WRLD - Hear Me Calling'
```

```python
from sklearn.metrics.pairwise import euclidean_distances

cluster = list(projection_m[projection_m['song']== nome_musica]['cluster_pca'])[0]
musicas_recomendadas = projection_m[projection_m['cluster_pca']== cluster][[0, 1, 'song']]
x_musica = list(projection_m[projection_m['song']== nome_musica][0])[0]
y_musica = list(projection_m[projection_m['song']== nome_musica][1])[0]

#distâncias euclidianas
distancias = euclidean_distances(musicas_recomendadas[[0, 1]], [[x_musica, y_musica]])
musicas_recomendadas['id'] = dados['id']
musicas_recomendadas['distancias']= distancias
recomendada = musicas_recomendadas.sort_values('distancias').head(10)
recomendada
```

## **Criando uma função para melhor utilização do algoritmo.**

```python
def musica(nome_musica):

  from sklearn.metrics.pairwise import euclidean_distances

  cluster = list(projection_m[projection_m['song']== nome_musica]['cluster_pca'])[0]
  musicas_recomendadas = projection_m[projection_m['cluster_pca']== cluster][[0, 1, 'song']]
  x_musica = list(projection_m[projection_m['song']== nome_musica][0])[0]
  y_musica = list(projection_m[projection_m['song']== nome_musica][1])[0]

  #distâncias euclidianas
  distancias = euclidean_distances(musicas_recomendadas[[0, 1]], [[x_musica, y_musica]])
  musicas_recomendadas['id'] = dados['id']
  musicas_recomendadas['distancias']= distancias
  recomendada = musicas_recomendadas.sort_values('distancias').head(10)
  return recomendada
```

```python
musica('Frank Ocean - White Ferrari')
```

```python
musica('Kanye West - Ghost Town')
```
